```python
!pip install faiss-gpu chromadb faiss-cpu datasets PyMuPDF pdfplumber sentence-transformers
import uuid
import os
import torch
import fitz
import pdfplumber
from datasets import Dataset
from unsloth import FastLanguageModel
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions

# --- Step 1: Load Model with QLoRA Optimization ---
print("Checking CUDA availability...")
if torch.cuda.is_available():
    print(f"CUDA Available: {torch.cuda.is_available()}")
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
else:
    print("No GPU detected! Check settings.")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="google/gemma-2-2b",
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=True,
    device_map="auto",
)

# --- Step 2: Initialize Vector Database for RAG ---
print("Initializing Vector Database...")
chroma_client = chromadb.PersistentClient(path="./rag_vector_db")
collection = chroma_client.get_or_create_collection(
    name="medical_docs",
    embedding_function=embedding_functions.SentenceTransformerEmbeddingFunction(
        model_name="sentence-transformers/all-mpnet-base-v2"
    )
)

# --- Step 3: Improved PDF Processing with Semantic Chunking ---
def semantic_chunking(text, max_length=500):
    """Split text into semantically meaningful chunks"""
    sentences = text.split('. ')
    chunks = []
    current_chunk = ""

    for sentence in sentences:
        if len(current_chunk) + len(sentence) < max_length:
            current_chunk += sentence + ". "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + ". "

    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

def process_pdf_to_chunks(pdf_path):
    """Extract and chunk PDF content"""
    try:
        # First try pdfplumber for better text extraction
        with pdfplumber.open(pdf_path) as pdf:
            full_text = ""
            for page in pdf.pages:
                text = page.extract_text()
                if text:
                    full_text += text + "\n"

        if len(full_text) < 100:
            doc = fitz.open(pdf_path)
            full_text = ""
            for page in doc:
                full_text += page.get_text("text") + "\n"

        return semantic_chunking(full_text)
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
        return []

# --- Step 4: Build Knowledge Base ---
def build_knowledge_base(pdf_paths):
    """Process PDFs and add to vector database"""
    for path in pdf_paths:
        chunks = process_pdf_to_chunks(path)
        if not chunks:
            continue

        embeddings = collection._embedding_function(chunks)
        ids = [str(uuid.uuid4()) for _ in chunks]

        collection.add(
            documents=chunks,
            embeddings=embeddings,
            ids=ids,
            metadatas=[{"source": path}] * len(chunks)
        )
    print(f"Added {len(chunks)} chunks from {len(pdf_paths)} files")

# --- Step 5: Create RAG-Augmented Training Data ---
def generate_rag_training_examples(base_dataset, num_examples=500):
    """Enhance training data with relevant context"""
    rag_examples = []

    for example in base_dataset.select(range(min(num_examples, len(base_dataset)))):
        # Retrieve relevant context
        query = example["Question"]
        results = collection.query(
            query_texts=[query],
            n_results=3
        )
        context = "\n".join(results["documents"][0])

        formatted_example = (
            "### Medical Context:\n" + context + "\n\n" +
            "### Question:\n" + example["Question"] + "\n\n" +
            "### Response:\n" + example["Response"] + tokenizer.eos_token
        )
        rag_examples.append(formatted_example)

    return Dataset.from_dict({"text": rag_examples})

# --- Step 6: Apply LoRA & Freeze Weights ---
print("Applying LoRA and freezing weights...")
# model = FastLanguageModel.get_peft_model(
#     model,
#     r=32,  # Increased from 16 for better adaptation
#     target_modules=[
#         "q_proj", "k_proj", "v_proj", "o_proj",
#         "gate_proj", "up_proj", "down_proj"
#     ],
#     lora_alpha=32,
#     lora_dropout=0.1,  # Added dropout for regularization
#     bias="none",
#     use_gradient_checkpointing=True,  # Enable for memory efficiency
#     random_state=3407,
# )

print("Applying LoRA...")
model = FastLanguageModel.get_peft_model(
    model,
    r=32,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ],
    lora_alpha=32,
    lora_dropout=0.1,
    bias="none",
    use_gradient_checkpointing=True,
    random_state=3407,
)

print("LoRA applied successfully!")

def freeze_layers(model, freeze_ratio=0.7):
    named_params = list(model.named_parameters())
    total_layers = len(named_params)
    trainable_layers = int(total_layers * (1 - freeze_ratio))

    for i, (name, param) in enumerate(named_params):
        if param.is_floating_point():
            param.requires_grad = (i >= (total_layers - trainable_layers))
            status = "Trainable" if param.requires_grad else "Frozen"
            print(f"{status}: {name}")

    print(f"\nSummary: Frozen {total_layers - trainable_layers}/{total_layers} layers")
def safe_freeze_layers(model, freeze_ratio=0.7):
    """Freeze layers while respecting LoRA parameters"""
    params_to_freeze = [
        (name, param) for name, param in model.named_parameters()
        if param.requires_grad
        and "lora_" not in name
        and "norm" not in name
        and param.is_floating_point()
    ]

    num_to_freeze = int(len(params_to_freeze) * freeze_ratio)

    for name, param in params_to_freeze[:num_to_freeze]:
        param.requires_grad = False
        print(f"Froze: {name}")

    print(f"\nFroze {num_to_freeze}/{len(params_to_freeze)} trainable parameters")


# freeze_layers(model)
safe_freeze_layers(model)

# --- Step 7: Training with RAG-Enhanced Data ---
from transformers import TrainingArguments
from trl import SFTTrainer

# dataset = load_dataset("FreedomIntelligence/medical-o1-reasoning-SFT", "en", split="train[0:500]")

print("build knowledge base")
pdf_paths = ["/content/heart_1.pdf", "/content/heart_2.pdf", "/content/heart_3.pdf"]
build_knowledge_base(pdf_paths)

all_chunks = []
for path in pdf_paths:
    chunks = process_pdf_to_chunks(path)
    all_chunks.extend(chunks)

print("Create QA pairs from chunks" )
train_texts = []
for chunk in all_chunks[:500]:
    train_texts.append(
        "### Medical Context:\n" + chunk + "\n\n" +
        "### Question:\nWhat is the key information in this medical text?\n\n" +
        "### Response:\n" + chunk.split('.')[0] + tokenizer.eos_token
    )

print("Train the")
train_dataset = Dataset.from_dict({"text": train_texts})

training_args = TrainingArguments(
    per_device_train_batch_size=2,
    gradient_accumulation_steps=2,
    max_steps=300,
    learning_rate=1e-4,
    fp16=not torch.cuda.is_bf16_supported(),
    bf16=torch.cuda.is_bf16_supported(),
    output_dir="outputs",
    logging_steps=10,
    evaluation_strategy="steps",
    eval_steps=50,
    save_steps=100,
    report_to="none",
    warmup_ratio=0.1,
)

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    dataset_text_field="text",
    max_seq_length=2048,
    args=training_args,
)

print("Starting training...")
trainer.train()

# --- Step 8: Save Model ---
repo_name = "GVAmaresh/gemma-2b-medical-rag"
model.push_to_hub(repo_name)
tokenizer.push_to_hub(repo_name)

print("Model fine-tuned and uploaded successfully!")
```
